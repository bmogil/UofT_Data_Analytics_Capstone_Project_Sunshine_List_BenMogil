# Create 1 Consolidated Ontario Wages DataFrame

```python
#Create Wage DataFrame - combine Male data
Ontario_Wages= wages_male1_df.append(wages_male2_df, ignore_index = True)
# Ontario_Wages.head()
Ontario_Wages.tail()
```

# Cleaning of the Statics Canada - CPI

```python
#File to Load Ontario Males 25-54 wages for 1997 to 2021
cpi_data_to_load = "Raw Data CSV files/OntarioCPI_databaseLoadingData.csv"
```

```python
# Create a DataFrame for Ontario Males 25-54 wages for 1997 to 2021
cpi_df = pd.read_csv(cpi_data_to_load)

#display the data table for preview
cpi_df.tail()
```

```python
#Review data types
cpi_df.dtypes
```

```python
# find null values
for column in cpi_df.columns:
    print(f"Column {column} has {cpi_df[column].isnull().sum()}null values")
```

```python
# Drop null value, not required columns Status, Symbol and Terminated
cpi_df.drop(columns=['UOM','DECIMALS','DGUID','UOM_ID','SCALAR_FACTOR', 'SCALAR_ID','VECTOR','COORDINATE','STATUS','SYMBOL','TERMINATED'], inplace=True)
```

```python
# Rename the columns in the sunshine_data DataFrame.### changed to match 2020 data
cpi_df.rename({'REF_DATE':'year',
                         'GEO':'province',
                         'Products and product groups':'product_groups',
                         'VALUE':'CPI'
                    }, axis='columns', inplace=True)
```

```python
cpi_df.columns
```

```python
# review types of product_groups for standardization with sunshine list
cpi_df['product_groups'].unique()
```

```python
#Keep only All-items category
remove = ['Food', 'Shelter',
       'Household operations, furnishings and equipment',
       'Clothing and footwear', 'Transportation', 'Gasoline',
       'Health and personal care', 'Recreation, education and reading',
       'Alcoholic beverages, tobacco products and recreational cannabis',
       'All-items excluding food and energy',
       'All-items excluding energy', 'Energy', 'Goods', 'Services']

cpi_df=cpi_df.drop(cpi_df[cpi_df['product_groups'].isin(remove)].index)
print(cpi_df.shape)                  
```

```python
cpi_df.dtypes
```

```python
# Export to csv 
cpi_df.to_csv("Resources/Clean_CPI.csv", index=False)
cpi_df.head()
```
